```python
import xlrd  
from locationiq.geocoder import LocationIQ
from math import radians,sin,cos,acos
import json

class Node:
       def __init__(self,state,cost,parent):
            self.state=state
            self.cost=cost
            self.Parent=parent
        
```

```python
        
def explorenode(src,cost):
        loc = ("/home/shankar/Downloads/Indian_capitals.xlsx") 
        wb = xlrd.open_workbook(loc) 
        sheet = wb.sheet_by_index(0)
        b=sheet.nrows
        a=sheet.ncols
        array=[]
        for i in range(0,b):
            for j in range(0,a):
                    if sheet.cell_value(i,j)==src:
                        if(j!=1):
                            state=sheet.cell_value(i,1).lstrip()
                            cost=sheet.cell_value(i,2)
                            node=Node(state,cost,src)
                            array.append(node)

                        if(j==1):
                            state=sheet.cell_value(i,0).lstrip()
                            cost=sheet.cell_value(i,2)
                            node=Node(state,cost,src)
                            array.append(node)
        return array
    
    
```

```python
def hn(src,dest):
    geocoder = LocationIQ("df1cea1ac0972a")
    x=geocoder.geocode(src)[0]
    y=geocoder.geocode(dest)[0]
    slat=radians(float(x['lat']))
    slon=radians(float(x['lon']))
    elat=radians(float(y['lat']))
    elon=radians(float(y['lon']))
    dist=6371.01*acos(sin(slat)*sin(elat)+cos(slat)*cos(elat)*cos(slon-elon))
    return dist
```

```python

def dfs(src,goal):
    explored={'0'}
    stack=[]
    seq=[]
    x=Node(src,None,0)
    stack.append(x)
    explored.add(x.state)
    if src==goal:
        return
    while stack:
        src=stack.pop()
        seq.append(src.state)
        if src.state==goal:
            return seq
        result=explorenode(src.state,0)
        for i in range(0,len(result)):
             if result[i]  not in explored:
                stack.append(result[i])
                explored.add(result[i].state)
```

```python
def bfs(src,goal):
    visited={'0'}
    Queue=[]
    seq=[]
    x=Node(src,0,'None')
    Queue.append(x)
    if src==goal:
        return 0
    visited.add(x.state)
    while Queue:
        src=Queue.pop(0)
        seq.append(src.state)
        if src.state==goal:
            return seq
        result=explorenode(src.state,0)
        for i in range(0,len(result)):
             if result[i].state  not in visited:
                Queue.append(result[i])
                visited.add(result[i].state)
```

```python
def bdbfs(src,goal):
    visit1=set()
    visit2=set()
    front=[]
    back=[]
    actions=[]
    node1=Node(src,'None',0)
    front.append(node1)
    node2=Node(goal,'None',0)
    back.append(node2)
    visit1.add(node1.state)
    visit2.add(node2.state)
    array1=[]
    array2=[]
    while len(front)!=None and len(back)!=None:
        if set(visit1) & set(visit2):
           for element in visit1:
               if element in visit2:
                   array1.append(element)
                   array1=array1+array2
                   return array1
            
        node1=front.pop(0)                                                               
        array1.append(node1.state)
        actions=explorenode(node1.state,0)
        for i in range(0,len(actions)):
             if actions[i].state  not in visit1:
                front.append(actions[i])
                visit1.add(actions[i].state)
                
        node2=back.pop(0)
        array2.append(node2.state)
        actions=explorenode(node2.state,0)
        for i in range(0,len(actions)):
             if actions[i].state  not in visit2:
                back.append(actions[i])
                visit2.add(actions[i].state)


```
